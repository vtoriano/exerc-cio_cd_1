## Passo 0 — Importar pandas e carregar os dados

```python
import pandas as pd
df = pd.read_parquet ("https://drive.google.com/uc?export=download&id=19ydA23-eboyiKxN29JL1EQWW4CplTiKg")
```

## 1) Filtre apenas as emendas do ano de **2022** e mostre as 30 primeiras linhas. Em seguida responda: Quanta emendas foram registradas em 2022?

```python
df_2022 = df[df['Ano da Emenda'] == 2022]
print (df_2022.shape[0])
```

## 2) Filtre as emendas da região **Norte** que tenham `Valor Empenhado` maior que **500.000** e mostre as 5 primeiras linhas. Em seguida, responda: quantas emendas atendem a esses duas condições ao mesmo tempo?

```python
df_norte = df[(df['Região'] == 'Norte') & (df['Valor Empenhado'] > 50000)]
df_norte.head()
```

## 3) Filtre as emendas que são do estado de **SÃO PAULO** ou do estado do **RIO DE JANEIRO** e mostre quantas linhas cada estado possui. Em seguida, responda: Qual dos dois estados possui mais emendas e quantas são?

```python
sp_rj = df[df['UF'].isin(['SÃO PAULO', 'RIO DE JANEIRO'])]
print(sp_rj['UF'].value_counts())
```

## 4) Na coluna `Código da Emenda`, substitua todos os valores `"Sem informação"` por `"Não disponível"`. Mostre a contagem de valores únicos da coluna antes e depois da substituição.

```python
df['Código da Emenda'] = df['Código da Emenda'].replace('sem informação', 'não disponivel')
print(df['Código da Emenda'].value_counts())
```

## 5) Na coluna `Nome Função`, transforme todos os valores para que fiquem em letras maiúsculas. Confirme a alteração mostrando os valores únicos da coluna.

```python
df['Nome Função'] = df['Nome Função'].str.upper()
print(df['Nome Função'].unique())
```

## 6) Verifique quantas linhas duplicadas existem no dataset. Em seguida, remova as duplicatas e mostre quantas linhas sobraram.

```python
print(f'Linhas dupicadas: {df.duplicated().sum()}')
df = df.drop_duplicates()
print(f'Linhas após remoção: {len(df)}')
```

## 7) Remova duplicatas considerando apenas as colunas `Código da Emenda` e `Município`. Quantas linhas sobraram? O resultado é diferente de quando removemos duplicatas do DataFrame inteiro?

```python
df_novo = df.drop_duplicates(subset=['Código da Emenda', 'Município'])
df_novo.shape[0]
```

## 8) Identifique quais colunas possuem valores nulos e quantos nulos cada uma tem. Em seguida, responda: qual é a coluna que possui maior número de valores nulos e quantos são?

```python
df_nulos = df.isnull().sum()
df_nulos
```

## 9) Preencha os valores nulos da coluna `codigo do autor` com **-1** e os valores nulos da coluna `nome do autor da emenda` com "desconhecido" e confirme que não restaram nulos nessa coluna.

```python
df['Código do Autor da Emenda'] = df['Código do Autor da Emenda'].fillna(-1)
df['Nome do Autor da Emenda'] = df['Nome do Autor da Emenda'].fillna('desconhecido')
```

## 10) Verifique o tipo da coluna `Ano da Emenda`. Converta-a para **string** (`str`). Mostre o tipo antes e depois da conversão.

```python
df['Ano da Emenda'] = df['Ano da Emenda'].astype(str)
```

## 11) Converta a coluna `Valor Empenhado` para o tipo **float** (caso ainda não seja). Mostre o tipo após a conversão.

## 12) Calcule o **total** de `Valor Pago` por **Região**. Quais são as duas regiões que mais receberam recursos?

```python
df_regiao = df.groupby('Região')['Valor Pago']
df_regiao
```

## 13) Calcule a **quantidade** de emendas por **Nome Função**. Mostre as 5 funções com mais emendas.

```python
em_nome_func = df.groupby('Nome Função')['Código da Emenda'].count().sort_values(ascending=False)
em_nome_func.head()
```

## 14) Crie uma nova coluna chamada `Saldo a Pagar`. Ela deve ser o resultado da subtração entre a coluna `Valor Empenhado` e a coluna `Valor Pago`. Mostre as 5 primeiras linhas apenas com essas três colunas.

```python
df['Saldo a Pagar'] = df['Valor Empenhado'] - df['Valor Pago']
df['Saldo a Pagar'].head()
```
